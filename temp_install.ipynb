# Install Supabase in the project environment
This notebook checks the current Python interpreter and installs the required Supabase package into the active environment.

```python
import sys
import subprocess
print('Python executable:', sys.executable)
print('Python version:', sys.version)

# Install a compatible Supabase version and its dependencies
result = subprocess.run([sys.executable, '-m', 'pip', 'install', 'supabase==1.2.0'], capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
print('Exit code:', result.returncode)
```
